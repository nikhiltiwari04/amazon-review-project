```python
class Sentiment:
    NEGATIVE = 'NEGATIVE'
    NUETRAL = 'NEUTRAL'
    POSITIVE = 'POSITIVE'

class Review:
    def __init__(self, text, score ):
        self.text = text
        self.score = score
        self.sentiment = self.get_sentiment()

    def get_sentiment(self):
        if self.score <= 2:
            return Sentiment.NEGATIVE
        elif self.score ==3:
            return Sentiment.NUETRAL
        elif self.score >= 4:
             return Sentiment.POSITIVE
```

```python
import json

file_name = 'D:/Coding/Ml/books_small.json'
reviews = []

with open(file_name, 'r', encoding='utf-8') as f:
    for line in f:
        review_json = json.loads(line)
        review_obj = Review(review_json['reviewText'], review_json['overall'])
        reviews.append(review_obj)

# Display one review
reviews[1].sentiment
```
